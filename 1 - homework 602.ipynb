````python
#Does each country have a row for every year present?

result = []
for country in df.country.unique():
    cntry_df = df[df['country'] == country]
    if len(cntry_df.year.unique()) != len(df.year.unique()):
        result.append(country)
        
print("Below are the list of countries where there isn't a row for every year present:\n", result)
````

## Question 2: Wide data isn't usually helpful.
Convert this to a long format with the following columns:  
1. country  
2. year  
3. diagnosis method: mapping should be: 

```
  {'newsp':'Pulmonary Smear (+)'
  ,'newsn':'Pulmonary Smear (-)'
  ,'newep':'Extrapulmonary'
  ,'newrel':'Relapse'}   
 ```

4. gender: male or female.   
5. age: lower_age - higher_age, e.g., 0-14  
6. number of cases  

__Note__: You can use `replace` (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.replace.html)

````python
# Dropping all rows where there no non-null values

new_df = df[list_of_cols].dropna(thresh=1)
new_df = pd.merge(df, new_df, how='inner', on=list_of_cols)
new_df = new_df.drop_duplicates()
new_df
````

````python
dfLong = (
    new_df.melt(id_vars=['country', 'year'], var_name='diagnosis')
)

dfLong.head()
````

````python
dfLong['gender'] = [x.replace('new_', 'new').split('_')[1] for x in dfLong['diagnosis']]
dfLong['diagnosis'] = [x.replace('new_', 'new').split('_')[0] for x in dfLong['diagnosis']]
dfLong['age'] = [x[1:2]+'-'+x[2:] if len(x[1:])==3 else x[1:3]+'-'+x[3:] for x in dfLong['gender']]
dfLong['age'] = [x[:-1] if x[-1]=='-' else x for x in dfLong['age']]
dfLong['gender'] = ['male' if x[0:1]=='m' else 'female' for x in dfLong['diagnosis']]
dfLong.rename(columns={"value": "no_of_cases"}, inplace=True)

dfLong
````
